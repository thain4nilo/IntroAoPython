# While
Laço que se assemelha ao if, mas caso a condição seja satisfeita ele entra em loop, ou seja, repete a operação identada.

**atenção:** precisa inserir uma forma de sair do loop, caso contrario o programa só encerrará com um 'stop' na IDE



#### Exercícios - While

````python
#Escreva um programa que leia como entrada uma taxa de juros de ponto flutuante
#(por exemplo, 0.06, que corresponde a 6% de taxa de juros). Seu programa deverá
#calcular e imprimir quanto tempo (em anos) será necessário para que um
#investimento duplique de valor. 

# Nota: O número de anos necessário para que um investimento duplique
# não depende do valor do investimento inicial.

juros = float(input("Qual a taxa de juros? "))

investimento = 1
valor = investimento
anos = 0
while valor <= (investimento*2):
    anos += 1
    valor *= (1 + juros)
    
print(f'Seu investimento duplicará em {anos} anos')
````

# Comando break, continue e pass
O comando **break**:
* É usado no corpo de um laço
* Quando executado, interrompe o *laço* atual
* A execução continua com a linha logo após o corpo do laço

O comando **continue**:
* É usado no corpo de um laço
* Quando executado, interrompe a *iteração* atual
* A execução continua com a próxima iteração do laço

O comando **pass**:
* É usado para substituir um corpo, Cada comando def, if ou laço (for ou while) deve ter um corpo, do contrário haveria um erro de sintaxe.
* Não executa nenhuma ação
* Usado geralmente quando o código está em desenvolvimento
* Simplesmente passa e ainda assim é um corpo válido

#### Exercícios - break, continue e pass

````python
table = [
[2, 3, 0, 6],
[0, 3, 4, 5],
[4, 5, 6, 0]]

print("break")
for row in table:
    for num in row:
        if num == 0:
            break #quando chegar em zero ele n vai printar o restante da linha, ou seja vai sair deste laço de for
        print(num, end=' ')
    print()

print()
print("continue")

for row in table:
    for num in row:
        if num == 0: 
            continue #quando chegar em zero ele n vai printar o 0, ou seja vai pular para o proximo valor de num
        print(num, end=' ')
    print()

print()
print("pass")

for row in table:
    for num in row:
        if num == 0: 
            pass #não faz nada
        print(num, end=' ')
    print()
````

````python
# Escreva um programa que:
# Exibe três opções ao usuário: 1- entrar, 2- exibir tentativas e 3-sair
# O sistema deve pedir ao usuário para escolher uma opção
# Na opção 1, o usuário deverá digitar uma senha. Se esta senha for igual a frase secreta (pythonEhLegal), então o usuário entra no sistema e zera as tentativas. Caso contrário, o número de tentativas é incrementado. Note que o usuário só pode tentar no máximo 3 vezes.
# Na opção 2, o sistema exibe quantas tentativas foram feitas até o momento
# Na opção 3, o sistema deve finalizar

tentativas = 0
acertou = False
senha = 'pythonEhLegal'
while tentativas < 3 and acertou == False:
    opcao = int(input('Escolha uma opcao: 1- entrar, 2- exibir tentativas e 3-sair '))

    if opcao == 1:
        s = input('Digite uma senha: ')
        if s == senha:
            tentativas = 0
            acertou = True
            print('Senha correta')
        else:
            tentativas += 1
    elif opcao == 2:
        print('Ja foram realizadas %d tentativas' % tentativas)
    elif opcao == 3:
        break
````

````python
# Implemente a função rlookup ( ), que oferece o recurso de pesquisa reversa de uma agenda de telefones.
# Sua função aceita, como entrada, um dicionário representando uma agenda de telefones. No dicionário, 
# os números de telefone (chaves) são mapeados para indivíduos (valores). Sua função deverá oferecer uma
# interface de usuário simples, por meio da qual um usuário pode inserir um número de telefone e obter o
# nome e sobrenome do indivíduo atribuído a esse número. 

def rlookup(agenda_r):
    numero = str(input('Digite numero do telefone no formato (xxx)xxx-xx-xx: '))

    if numero in agenda_r:
        print(agenda_r[numero])
    else:
        print('O numero informado nao esta em uso.')

agenda_r = {'(123)456-78-90' : ['Anna', 'Karenina'],
            '(901)234-56-78' : ['Yu', 'Tsun'],
            '(321)908-76-54' : ['Hans', 'Castorp']}

rlookup(agenda_r)
````

````python
#Implemente uma função wordcount() que:
# - Recebe como entrada um text (string) e
# - Imprime a frequência com que cada palavra aparece no texto.
# Obs: Assuma que não há pontuação no texto.

text = "quem tem boca vai a roma mas o rato roeu a roupa do rei de roma"

def wordcount(texto):
    palavrasTexto = texto.split()
    dicPalavras = {}
    for palavra in palavrasTexto:
        if palavra in dicPalavras:
            dicPalavras[palavra] += 1
        else:
            dicPalavras[palavra] = 1
    
    for palavra in dicPalavras:
        print(f'{palavra} - foi encontrada {dicPalavras[palavra]} vez(es)')
            

wordcount(text)
````

````python
# Escreva uma função chamada lookup() para implementar
# pesquisas na agenda telefônica. Essa função aceita como entrada
# um dicionário da seguinte forma:

agenda = {('Anna','Karenina'):'(123)456-78-90',
          ('Yu', 'Tsun'):'(901)234-56-78',
          ('Hans', 'Castorp'):'(321)908-76-54'}
           
# A função deve ter um serviço contínuo que pede ao usuário nome e
# sobrenome e retorna o número, caso o contato estiver na agenda.

def lookup(agenda):
    while True:
        nome = input("Nome: ")
        if nome != "":
            sobrenome = input("Sobrenome: ")
            print("Telefone: ",agenda[nome,sobrenome])
        else:
            print("encerrando programa")
            break
           
lookup(agenda)
````

# Arquivos


### Modos de abertura
**r** - Modo de leitura (padrão)

**w** - Modo de gravação; se o arquivo já existir, seu conteúdo é apagado

**a** - Modo de acréscimo; o conteúdo é acrescentado ao final do arquivo

**r+** - Modo de leitura e gravação (pouco usado)

**t** - Modo de texto (padrão)

**b** - Modo binário

### Métodos


### Padrões de leitura
1. Ler o conteúdo do arquivo em uma string


````python
def numChars(filename):
    'retorna o numero de caracteres em um arquivo'
    infile = open(filename)
    content = infile.read() #Lê todo o conteudo em uma só string
    infile.close () #Importante fechar o arquivo logo após terminar de usar
    
    return len(content)

print(numChars("example.txt"))
````

2. Ler o conteúdo do arquivo em uma lista de palavras

````python
def numWords(filename):
    'retorna o numero de palavras em um arquivo'
    infile = open(filename)
    content = infile.read()  #Lê todo o conteudo em uma só string
    infile.close () #Importante fechar o arquivo logo após terminar de usar
    wordList = content.split()
    return len(wordList)

print(numWords("example.txt"))
````

3. Ler o conteúdo do arquivo em uma lista de linhas

````python
def numLines(filename):
    'retorna o numero de linhas em um arquivo'
    infile = open(filename)
    lineList = infile.readlines()  #Lê cada linha, retornando uma lista de strings 
    infile.close () #Importante fechar o arquivo logo após terminar de usar

    return len(lineList)

print(numLines("example.txt"))
````

#### Exercícios - arquivos

````python
# Implemente a função meuGrep() que:
# 1. Toma como entrada duas strings: um nome de arquivo e uma string alvo
# 2. Exibe cada linha do arquivo que contém a string alvo

def meuGrep(arquivo,palavra):
    infile = open(arquivo)
    for line in infile:
        if palavra in line.split():
            print(line,end='')
    infile.close()
        
meuGrep("example.txt","rato")
````
